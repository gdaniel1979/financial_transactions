## Cleaning cards_data.csv in BASH

#### cards_data_cleaning.sh

```python
#!/bin/bash

# Input CSV
INPUT="cards_data.csv"
OUTPUT="cards_data_clean.csv"

# Removing $ signs, converting data formats
sed 's/\$//g' "$INPUT" \
| awk -F',' 'BEGIN{OFS=","}
NR==1 {print; next}
{
    split($6, e, "/")
    $6 = e[2] "-" sprintf("%02d", e[1]) "-01"

    split($11, a, "/")
    $11 = a[2] "-" sprintf("%02d", a[1]) "-01"

    print
}' > "$OUTPUT"

echo "Done: $OUTPUT"
```

```python
chmod +x cards_data_cleaning.sh
```

```python
./cards_data_cleaning.sh
```

## Cleaning users_data.csv in BASH

```python
#!/bin/bash

# Input CSV
INPUT="users_data.csv"
OUTPUT="users_data_clean.csv"

# Removing $ signs
sed 's/\$//g' "$INPUT" > "$OUTPUT"

echo "Done: $OUTPUT"
```

```python
chmod +x users_data_cleaning.sh
```

## Cleaning transactions_data.csv in BASH

```python
#!/bin/bash

# Input CSV
INPUT="transactions_data.csv"
OUTPUT="transactions_data_clean.csv"

# Cleaning
# Removing $ signs
# Removing .0 from ZIP column (58523.0 -> 58523)

sed 's/\$//g' "$INPUT" | sed 's/\.0,/,/g' | sed 's/\.0$/ /' \
> "$OUTPUT"

echo "Done: $OUTPUT"
```

```python
chmod +x transactions_data_cleaning.sh
```

```python
./transactions_data_cleaning.sh
```
